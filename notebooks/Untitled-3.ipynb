```python
import torch 
import torch.nn as nn 
import torch.optim as optim 

```

```python
class MedoNN(nn.Module):
    def _init_(self):
        super(MedoNN, self )._init_()
        self.fc1= nn.Linear(784,128)
        self.relu=nn.ReLU()
        self.fc2 = nn.Linear(128,10)

    def forward(self,x):
        x=x.view(-1,784)
        x=self.fc1(x)
        x=self.relu(x)
        x=self.fc2(x)
        return x 
    
```

```python
model = MedoNN()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```
